# Create First Iteration of data:

```python
import pandas as pd
import numpy as np

from pathlib import Path
```

```python
# 2 reports for base result:
base_reports = ['41866595.xlsx', '41866596.xlsx']
df_1 = pd.read_excel('./data/source/'+base_reports[0])
df_1['report_number'] = '41866595'
df_2 = pd.read_excel('./data/source/'+base_reports[1])
df_2['report_number'] = '41866596'
print(f'df_1.shape={df_1.shape}')
print(f'df_2.shape={df_2.shape}')
df_result = df_1.copy()
df_result = df_result.drop(df_result.index)
# concat 1st 2 files:
df_result = pd.concat([df_1, df_2], ignore_index=True)
print(f'df_result.shape={df_result.shape}')
# save to csv:
df_result.to_csv('./data/result/df_result.csv', index=False, sep='|')
```

```python
reports_uploaded = list(df_result['report_number'].unique())
reports_uploaded
```

```python
def check_what_reports_are_new(reports_uploaded):
    folder_path = './data/source/'
    folder = Path(folder_path)
    for file in folder.iterdir():
        if file.is_file():
            file_name = file.name
            report_name = int(file_name.replace('.xlsx', ''))
            # check whether report number already uploaded:
            new_reports = []
            if report_name not in reports_uploaded:
                print(f'Found new report: {report_name}')
                new_reports.append(report_name)
    return new_reports

# add to datasource:
def add_to_datasource(new_reports):
    folder_path = './data/source/'
    for report_name in new_reports:
        file_name = report_name + '.xlsx'
        df_report = pd.read_excel(folder_path+file_name)
        df_report['report_number'] = report_name        
        # concat to main file:
        df_result = pd.concat([df_result, df_report], ignore_index=True)
```

```python
df_111 = pd.read_excel('./data/source/111.xlsx')
df_111['report_number'] = 111
df_111.head(1)
```

```python
def check_columns(df_1, df_2, report_name):
    # all columns for both dfs:
    main_columns = list(df_1.columns)
    new_columns = list(df_2.columns)
    # lists for absent columns:
    main_columns_absent = []
    new_columns_absent = []
    # filling list of missing columns:
    for main_column in main_columns:
        if main_column not in new_columns:
            main_columns_absent.append(main_column)
    for new_column in new_columns:
        if new_column not in main_columns:
            new_columns_absent.append(new_column)
    # print response:
    if len(main_columns_absent) == 0 and len(new_columns_absent) == 0:
        print('1')
        print('Все Ок набор колонок совпадает')
    else:
        if len(main_columns_absent) > 0 and len(new_columns_absent) > 0:
            print('2')
            print(f'В новом файле: {report_name} нет колонок: {main_columns_absent}')
            print(f'В новом файле: {report_name} появились новые колонки: {new_columns_absent}')
        elif len(main_columns_absent) > 0:
            print('3')
            print(f'В новом файле: {report_name} нет колонок: {main_columns_absent}')
            print('Удалить эти колонки из источника данных?')
        else:
            print('4')
            print(f'В новом файле: {report_name} появились новые колонки: {new_columns_absent}')
    return main_columns_absent, new_columns_absent

# use search function:
main_columns_absent, new_columns_absent = check_columns(df_result, df_111)
print(main_columns_absent, new_columns_absent)
```
